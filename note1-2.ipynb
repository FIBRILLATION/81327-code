## 基本的なライブラリの読み込みパターン
*   **import xxx xxx**<BR>
library をすべて読み込み、xxx.関数名 といった形で利用する。
*   **import xxx.a_func as b_func**<BR>
xxx library の a_func だけを b_func という別名をつけて読み込む。名前の衝突の防止や短縮型として利用する。
*   **from xxx import a_func**<BR>
xxx library の a_func だけ読み込み、a_func で利用する。from をつけると xxx は指定せず利用できる。

```python
# シグモイド関数を例に
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-10, 10, 0.1) # 等差数列を生成する
y = sigmoid(x)

print (type(y))
print (type(y[0]), y[0])
print (y)
```

```python
## リストからnp.arrayへの型変換

data0 = range(10) # 注意

data1 = []
for x in range(10):
    data1.append(x)

data2 = [x for x in range(10)] # Python 特有の書き方

data3 = np.array(range(10)) # range => numpy.ndarray 変換

data4 = np.array(data2) # list => numpy.ndarray 変換

print (type(data0), data0)
print (type(data1), data1)
print (type(data2), data2)
print (type(data3), data3)
print (type(data4), data4)

```

```python
x = np.arange(0, 10, 0.1) # 等差数列を生成する
plt.plot(x, np.sin(x))
plt.show()
# 画像のダウンロードは、略式では、画像をマウスで右クリックしてもできます。
```

```python
# 2つのグラフを1枚にまとめて表示し、legendを追加
plt.plot(x, np.sin(x), label='Sine')
plt.plot(x, np.cos(x), label='Cosine')
plt.xlabel("x")
plt.legend()
plt.show()
```

```python
# 色の指定
plt.plot(x, np.sin(x - 0), color='blue')
plt.plot(x, np.sin(x - 1), color='g')
plt.plot(x, np.sin(x - 2), color='0.75')        # グレースケール between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3)) # RGB tuple, values 0 to 1
plt.show()

```

```python
# 線の種類の指定の省略表記
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
plt.show()

```

```python
# ヒストグラムの作成
x = np.random.randn(1000) # 平均0、標準偏差1の正規分布に従う乱数を1000個作成
plt.hist(x, color="green") 
```

```python
# 複数のヒストグラムの表示
x = np.random.normal(-3, 1, 1000) # 平均 -3 標準偏差 1 の正規分布に従う乱数
y = np.random.normal(3, 2, 1000)  # 平均 3 標準偏差 2  の正規分布に従う乱数

plt.hist(x, color="green", alpha=0.5)
plt.hist(y, color="orange", alpha=0.5)

```
